```python
coords = read_csv('transects_MENP.csv')
```

```python
#for coord in coords:
#utm.to_latlon(coords[0].easting, coords[0].northing, 54, 'H')
new_coords = []
for index, row in coords.iterrows():
    new_coords.append([utm.to_latlon(row.easting, row.northing, 54, 'H')[0],utm.to_latlon(row.easting, row.northing, 54, 'H')[1]])
new_coords
```

```python
with open('WGS84.csv', 'wb') as f:
    wr = csv.writer(f, quoting=csv.QUOTE_ALL)
    wr.writerow(['easting','northing'])
    for line in new_coords:
        wr.writerow(line)
```
